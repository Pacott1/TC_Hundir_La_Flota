```python
import numpy as np
import random
def crear_tablero(filas=10, columnas=10):
    # Crea un tablero, relleno con espacios.
    return np.full((filas, columnas), " ")

def posicionar_barco(tablero, posiciones):
    # Coloca un barco (O) en las posiciones indicadas.
    for (fila, col) in posiciones:
        tablero[fila, col] = "O"

tablero = crear_tablero()
tablero_rival = crear_tablero()
tablero_rival_2 = crear_tablero()
print("Tablero inicial:\n", tablero)
barco_2 = [(0,1), (1,1)]
barco_4 = [(1,3), (1,4), (1,5), (1,6)]
posicionar_barco(tablero_rival, barco_2)
posicionar_barco(tablero_rival, barco_4)
#posicionar_barco(tablero, barco_2)
#posicionar_barco(tablero, barco_4)
print("\nTablero con barcos colocados:\n", tablero)

def disparar(tablero_rival, tablero_rival_2, fila, col):
    """Devuelve True si acierta, False si falla, None si ya dispar√≥ ah√≠."""
    if tablero_rival[fila, col] == "O":
        tablero_rival[fila, col] = "X"
        tablero_rival_2[fila, col] = "X"
        print(f"¬°Impacto en ({fila}, {col})!")
        print(tablero_rival_2)
        return True
    elif tablero_rival[fila, col] == " ":
        tablero_rival[fila, col] = "-"
        tablero_rival_2[fila, col] = "-"
        print(f"Agua en ({fila}, {col}).")
        print(tablero_rival_2)
        return False
    else:
        print(f"Ya se dispar√≥ en ({fila}, {col}).")
        return None
    
def disparo_rival(tablero):
    filas, cols = tablero.shape
    while True:
        fila = random.randint(0, filas - 1)
        col = random.randint(0, cols - 1)

        if tablero[fila, col] in ["O", " "]:
            if tablero[fila, col] == "O":
                tablero[fila, col] = "X"
                print(f"¬°Impacto en ({fila}, {col})!")
                print(tablero)
                return True
            else:
                tablero[fila, col] = "-"
                print(f"Agua en ({fila}, {col}).")
                print(tablero)
                return False
            
def comprobar_derrota(tablero):                                     
    """Devuelve True si no quedan barcos ('O') en el tablero."""
    return not np.any(tablero == "O")

def mostrar_tablero(tablero, tablero_2):
    print(tablero, tablero_2)
```

```python
# ==============================
# PROGRAMA PRINCIPAL
# ==============================
while True:
    print("=== ¬°Bienvenido a Hundir la Flota! ===")
    nombre = input("Introduce tu nombre: ")

    # Crear tableros
    tablero_jugador = np.full((10, 10), " ")
    tablero_rival = np.full((10, 10), " ")
    tablero_rival_visible = np.full((10, 10), " ")

    # Colocar barcos (solo inputs del jugador)
    print(f"\nHola {nombre}, coloca tus barcos en el tablero.")
    print("Tienes:")
    print("- 4 barcos de 1 posici√≥n")
    print("- 3 barcos de 2 posiciones")
    print("- 2 barcos de 3 posiciones")
    print("- 1 barco de 4 posiciones")

    barcos = {1: 4, 2: 3, 3: 2, 4: 1}
    for tama√±o, cantidad in barcos.items():
        for n in range(cantidad):
            print(f"\nBarco {n+1} de tama√±o {tama√±o}")
            orientacion = "H"
            if tama√±o > 1:
                orientacion = input("Orientaci√≥n (H horizontal / V vertical): ").upper()
            try:
                fila, col = map(int, input("Introduce la posici√≥n inicial (fila,col): ").split(","))
            except:
                print("Formato incorrecto, usa 'fila,col' (ejemplo: 3,5).")
                continue
            print(f"‚Üí Guardado: Barco de tama√±o {tama√±o}, orientaci√≥n {orientacion}, en ({fila}, {col})")

    print("\nColocando barcos del rival aleatoriamente...")
    # colocar_barcos_aleatorios(tablero_rival)

    turno_jugador = True

    # ==============================
    # BUCLE DE PARTIDA
    # ==============================
    while True:
        print(tablero_jugador/n, tablero_rival_2)

        if turno_jugador:
            print(f"\nTurno de {nombre}:")
            try:
                fila, col = map(int, input("Introduce las coordenadas para disparar (fila,col): ").split(","))
            except:
                print("Formato incorrecto. Usa el formato fila,col (ejemplo: 3,5).")
                continue

            resultado = disparar(tablero_rival, tablero_rival_visible, fila, col)

            if resultado is True:
                print("üéØ ¬°Has acertado! Vuelves a disparar.")
            elif resultado is False:
                print("üåä Has fallado. Le toca a la m√°quina.")
                turno_jugador = False
            else:
                print("‚ö†Ô∏è Esa posici√≥n ya fue atacada, el turno pasa igualmente.")
                turno_jugador = False

            if comprobar_derrota(tablero_rival):
                print(f"üéâ ¬°{nombre} ha ganado! Todos los barcos enemigos han sido hundidos.")
                break

        else:
            print("\nTurno de la m√°quina:")
            resultado = disparo_rival(tablero_jugador)

            if resultado:
                print("üí• La m√°quina ha acertado y repite turno.")
                turno_jugador = False
            else:
                print("üíß La m√°quina ha fallado. Te toca a ti.")
                turno_jugador = True

            if comprobar_derrota(tablero_jugador):
                print("üíÄ ¬°La m√°quina ha ganado! Todos tus barcos han sido hundidos.")
                break

    # ==============================
    # FIN DE PARTIDA / REINICIO
    # ==============================
    print("\nPartida terminada.")
    opcion = input("¬øQuieres jugar otra vez? (s/n): ").lower()
    if opcion != "s":
        print("¬°Gracias por jugar! Hasta la pr√≥xima.")
        break
    else:
        print("\nReiniciando partida...\n")
```
